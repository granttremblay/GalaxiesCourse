You can use `astropy.io.fits` to read in the image data like this:

An image is just a two-dimensional histogram (!!). We can plot it up with `matplotlib.pyplot.imshow`

```python
fig, ax = plt.subplots(figsize=(12,12))

ax.imshow(f436_data, norm=LogNorm(), origin='lower', vmin=0.01, vmax=6.0, cmap='magma')

ax.set_xlim(1800, 2800)
ax.set_ylim(1800, 2800)
```

```python
fig, ax = plt.subplots(figsize=(12,12))

ax.imshow(m82_data, norm=LogNorm(), origin='lower', vmin=0.01, vmax=0.6,  cmap='viridis')

ax.set_xlim(2000, 10000)
ax.set_ylim(2000, 10000)
```

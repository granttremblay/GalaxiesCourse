```python
data_dir = '/Users/grant/Dropbox/Teaching/BannekerGalaxies/Data/'
mystery_data_dir = data_dir + 'mystery/'
```

```python
os.listdir(data_dir)
```

```python
f436_data = fits.getdata(data_dir + 'zwII96_wfc_f436.fits')
f814_data = fits.getdata(data_dir + 'zwII96_wfc_f814.fits')
```

```python
fig, ax = plt.subplots(figsize=(12,12))

ax.imshow(f436_data, norm=LogNorm(), origin='lower', vmin=0.01, vmax=6.0, cmap='magma')

ax.set_xlim(1800, 2800)
ax.set_ylim(1800, 2800)
```

```python
fig, ax = plt.subplots(figsize=(12,12))

ax.imshow(f814_data, norm=LogNorm(), origin='lower', vmin=0.1, vmax=6.0,  cmap='viridis')

ax.set_xlim(1800, 2800)
ax.set_ylim(1800, 2800)
```

```python
b_over_i = f436_data / f814_data

fig, ax = plt.subplots(figsize=(12,12))

ax.imshow(b_over_i, origin='lower', norm=LogNorm(), vmin=0.0001, vmax=2, cmap='plasma')

ax.set_xlim(1800, 2800)
ax.set_ylim(1800, 2800)
```
